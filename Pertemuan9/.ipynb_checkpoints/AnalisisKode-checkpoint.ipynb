## Python - Global Variables
### Global Variables

```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

### Analisis
Kode tersebut mendefinisikan sebuah fungsi myfunc() yang mencetak "Python is awesome". Fungsi ini menggabungkan string "Python is " dengan nilai dari variabel global x, yang berisi "awesome".

```python
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

### Analisis
Kode tersebut memiliki dua variabel x dengan lingkup berbeda. Variabel x di dalam fungsi bersifat lokal, sedangkan x di luar fungsi bersifat global.

## The global Keyword

```python
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

### Analisis
Kode tersebut menggunakan kata kunci global untuk menjadikan variabel x bersifat global di dalam fungsi. Setelah fungsi myfunc() dijalankan, variabel x akan berisi "fantastic"

```python
x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

### Analisis
Kode tersebut mendeklarasikan variabel global x dengan nilai awal "awesome", lalu fungsi myfunc() mengubah nilai global tersebut menjadi "fantastic". Setelah fungsi dijalankan, x bernilai "fantastic"

## Python Data Types
### Built-in Data Types

```python
Text Type:	str
Numeric Types:	int, float, complex
Sequence Types:	list, tuple, range
Mapping Type:	dict
Set Types:	set, frozenset
Boolean Type:	bool
Binary Types:	bytes, bytearray, memoryview
None Type:	NoneType
```

### Analisis
Teks tersebut menjelaskan berbagai tipe data bawaan dalam Python. Tipe teks diwakili oleh str untuk menyimpan data berupa string. Tipe numerik terdiri dari int, float, dan complex untuk bilangan bulat, desimal, dan kompleks. Tipe urutan atau sequence meliputi list, tuple, dan range, yang digunakan untuk menyimpan kumpulan data berurutan. Tipe pemetaan (dict) menyimpan data dalam bentuk pasangan kunci dan nilai. Tipe himpunan terdiri dari set dan frozenset, yang menyimpan data unik tanpa urutan. Selain itu, ada tipe boolean (bool) untuk menyimpan nilai benar atau salah, tipe biner (bytes, bytearray, memoryview) untuk menangani data dalam bentuk biner, dan tipe NoneType yang mewakili nilai kosong atau tidak ada.

### Getting the Data Type

```python
x = 5
print(type(x))
```

### Analisis
Kode tersebut mencetak tipe data dari variabel x. Karena x berisi angka 5, hasilnya adalah <class 'int'>, yang berarti tipe datanya adalah integer.

### Setting the Data Type

```python
x = "Hello World"	str	
x = 20	int	
x = 20.5	float	
x = 1j	complex	
x = ["apple", "banana", "cherry"]	list	
x = ("apple", "banana", "cherry")	tuple	
x = range(6)	range	
x = {"name" : "John", "age" : 36}	dict	
x = {"apple", "banana", "cherry"}	set	
x = frozenset({"apple", "banana", "cherry"})	frozenset	
x = True	bool	
x = b"Hello"	bytes	
x = bytearray(5)	bytearray	
x = memoryview(bytes(5))	memoryview	
x = None	NoneType
```

Setiap tipe data digunakan untuk tujuan yang berbeda, seperti menyimpan teks, angka, koleksi data, atau nilai logika.

### Python Assignment Operators
### Assignment Operators

```python
=	x = 5	x = 5	
+=	x += 3	x = x + 3	
-=	x -= 3	x = x - 3	
*=	x *= 3	x = x * 3	
/=	x /= 3	x = x / 3	
%=	x %= 3	x = x % 3	
//=	x //= 3	x = x // 3	
**=	x **= 3	x = x ** 3	
&=	x &= 3	x = x & 3	
|=	x |= 3	x = x | 3	
^=	x ^= 3	x = x ^ 3	
>>=	x >>= 3	x = x >> 3	
<<=	x <<= 3	x = x << 3	
:=	print(x := 3)	x = 3
print(x)
```

### Analisis
Teks tersebut menjelaskan berbagai operator penugasan (assignment operators) dalam Python, yang digunakan untuk melakukan operasi pada variabel dan memperbarui nilai variabel tersebut dalam satu langkah. Berikut adalah penjelasannya:

=: Penugasan dasar, misalnya x = 5, yang artinya memberikan nilai 5 ke variabel x.

+=: Penugasan dengan penambahan, misalnya x += 3 yang setara dengan x = x + 3.

-=: Penugasan dengan pengurangan, misalnya x -= 3 yang setara dengan x = x - 3.

*=: Penugasan dengan perkalian, misalnya x *= 3 yang setara dengan x = x * 3.

/=: Penugasan dengan pembagian, misalnya x /= 3 yang setara dengan x = x / 3.

%=: Penugasan dengan modulus, misalnya x %= 3 yang setara dengan x = x % 3.

//=: Penugasan dengan floor division, misalnya x //= 3 yang setara dengan x = x // 3.

**=: Penugasan dengan eksponensial, misalnya x **= 3 yang setara dengan x = x ** 3.

&=: Penugasan dengan operasi bitwise AND, misalnya x &= 3 yang setara dengan x = x & 3.

|=: Penugasan dengan operasi bitwise OR, misalnya x |= 3 yang setara dengan x = x | 3.

^=: Penugasan dengan operasi bitwise XOR, misalnya x ^= 3 yang setara dengan x = x ^ 3.

>>=: Penugasan dengan operasi bitwise right shift, misalnya x >>= 3 yang setara dengan x = x >> 3.

<<=: Penugasan dengan operasi bitwise left shift, misalnya x <<= 3 yang setara dengan x = x << 3.

:=: Walaupun operator ini sering disebut sebagai "walrus operator", pada contoh x := 3, ini menginisialisasi x dengan nilai 3 dan langsung mencetaknya. Fungsi ini berguna dalam ekspresi untuk assignment dan pengujian pada saat yang sama (biasanya digunakan dalam ekspresi kondisional atau loop).

### The Walrus Operator

```python
numbers = [1, 2, 3, 4, 5]
count = len(numbers)
if count > 3:
    print(f"List has {count} elements")

if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```

### Analisis
Kode tersebut menunjukkan dua cara untuk mengukur panjang daftar (numbers) dan mencetak hasilnya jika jumlah elemen lebih dari 3.

Pada bagian pertama, panjang daftar dihitung terlebih dahulu dengan len(numbers) dan disimpan dalam variabel count. Kemudian, jika count lebih dari 3, maka hasilnya dicetak.

Pada bagian kedua, walrus operator (:=) digunakan untuk menghitung panjang daftar dan langsung menyimpan hasilnya ke dalam variabel count di dalam kondisi if. Jika count lebih dari 3, maka hasilnya juga dicetak.

Kedua blok kode akan mencetak "List has 5 elements" karena panjang daftar numbers adalah 5, dan kondisi count > 3 terpenuhi di kedua kasus.

### Python Comparison Operators
### Comparison Operators

```python
==	Equal	x == y	
!=	Not equal	x != y	
>	Greater than	x > y	
<	Less than	x < y	
>=	Greater than or equal to	x >= y	
<=	Less than or equal to	x <= y
```

### Analisis
Teks tersebut menjelaskan berbagai operator perbandingan dalam Python yang digunakan untuk membandingkan dua nilai:

==: Menyatakan sama dengan, digunakan untuk memeriksa apakah dua nilai sama, misalnya x == y.

!=: Menyatakan tidak sama dengan, digunakan untuk memeriksa apakah dua nilai tidak sama, misalnya x != y.

>: Menyatakan lebih besar dari, digunakan untuk memeriksa apakah nilai di kiri lebih besar dari nilai di kanan, misalnya x > y.

<: Menyatakan lebih kecil dari, digunakan untuk memeriksa apakah nilai di kiri lebih kecil dari nilai di kanan, misalnya x < y.

>=: Menyatakan lebih besar atau sama dengan, digunakan untuk memeriksa apakah nilai di kiri lebih besar atau sama dengan nilai di kanan, misalnya x >= y.

<=: Menyatakan lebih kecil atau sama dengan, digunakan untuk memeriksa apakah nilai di kiri lebih kecil atau sama dengan nilai di kanan, misalnya x <= y.

Operator-operator ini menghasilkan nilai True atau False berdasarkan hasil perbandingan.

```python
x = 5
y = 3

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

### Analisis
Kode tersebut membandingkan nilai variabel x (5) dan y (3) menggunakan berbagai operator perbandingan, dan mencetak hasilnya:

x == y mengecek apakah x sama dengan y, hasilnya False.

x != y mengecek apakah x tidak sama dengan y, hasilnya True.

x > y mengecek apakah x lebih besar dari y, hasilnya True.

x < y mengecek apakah x lebih kecil dari y, hasilnya False.

x >= y mengecek apakah x lebih besar atau sama dengan y, hasilnya True.

x <= y mengecek apakah x lebih kecil atau sama dengan y, hasilnya False.

### Chaining Comparison Operators

```python
x = 5

print(1 < x < 10)

print(1 < x and x < 10)
```

### Analisis
Kode tersebut memeriksa dua kondisi yang serupa, namun dengan cara yang berbeda. Pertama, 1 < x < 10 adalah bentuk perbandingan berantai di Python yang memeriksa apakah x berada di antara 1 dan 10. Karena x adalah 5, hasilnya True. Kedua, 1 < x and x < 10 menggunakan operator and untuk memeriksa dua kondisi terpisah, yaitu apakah x lebih besar dari 1 dan apakah x lebih kecil dari 10. Karena keduanya benar, hasilnya juga True. Jadi, output dari kedua pernyataan tersebut adalah True.

### Python Logical Operators

```python
and 	Returns True if both statements are true	x < 5 and  x < 10	
or	Returns True if one of the statements is true	x < 5 or x < 4	
not	Reverse the result, returns False if the result is true	not(x < 5 and x < 10)
```

### Analisis
Operator and mengembalikan True jika kedua kondisi benar, or mengembalikan True jika salah satu kondisi benar, dan not membalikkan hasil kondisi (mengubah True menjadi False dan sebaliknya).

```python
x = 5

print(x > 0 and x < 10)
```

### Analisis
Kode tersebut memeriksa dua kondisi menggunakan operator and. Kondisi pertama, x > 0, dan kondisi kedua, x < 10, keduanya benar karena x adalah 5. Oleh karena itu, hasil yang dicetak adalah True.

### Python Bitwise Operators
### Bitwise Operators

```python
& 	AND	Sets each bit to 1 if both bits are 1	x & y	
|	OR	Sets each bit to 1 if one of two bits is 1	x | y	
^	XOR	Sets each bit to 1 if only one of two bits is 1	x ^ y	
~	NOT	Inverts all the bits	~x	
<<	Zero fill left shift	Shift left by pushing zeros in from the right and let the leftmost bits fall off	x << 2	
>>	Signed right shift	Shift right by pushing copies of the leftmost bit in from the left, and let the rightmost bits fall off	x >> 2
```

### Analisis
Teks tersebut menjelaskan beberapa operator bitwise dalam Python:

& (AND): Mengatur setiap bit menjadi 1 jika kedua bit yang dibandingkan adalah 1.

| (OR): Mengatur setiap bit menjadi 1 jika salah satu dari dua bit yang dibandingkan adalah 1.

^ (XOR): Mengatur setiap bit menjadi 1 jika hanya satu dari dua bit yang dibandingkan adalah 1.

~ (NOT): Membalikkan semua bit, mengubah 1 menjadi 0 dan sebaliknya.

<< (Zero fill left shift): Menggeser bit ke kiri dan mengisi posisi kosong dengan nol, sementara bit kiri terluar akan terbuang.

>> (Signed right shift): Menggeser bit ke kanan dan mengisi posisi kosong dengan salinan bit terluar kiri, sementara bit kanan terluar akan terbuang.

Operator-operator ini digunakan untuk manipulasi bit pada angka dalam format biner.

```python
print(6 & 3)
```

### Analisis
Kode tersebut menggunakan operator bitwise AND untuk membandingkan angka 6 dan 3. Dalam biner, 6 adalah 0110 dan 3 adalah 0011. Operator & hanya menghasilkan 1 jika kedua bitnya adalah 1, sehingga hasilnya adalah 2.

```python
print(6 | 3)
```

### Analisis
Kode tersebut menggunakan operator bitwise OR untuk membandingkan angka 6 dan 3. Dalam biner, 6 adalah 0110 dan 3 adalah 0011. Operator | menghasilkan 1 jika salah satu bitnya adalah 1.

```python
print(6 ^ 3)
```

### Analisis
Kode tersebut menggunakan operator bitwise XOR untuk membandingkan angka 6 dan 3. Dalam biner, 6 adalah 0110 dan 3 adalah 0011. Operator ^ menghasilkan 1 jika hanya satu bit yang 1 pada posisi tersebut.

Karena tugas telah diserahkan, program mencetak "Pass with good standing".

## Python Match
### The Python Match Statement

```python
match expression:
  case x:
    code block
  case y:
    code block
  case z:
    code block
```

### Analisis
Teks tersebut menunjukkan penggunaan match-case expression, yang diperkenalkan di Python 3.10, sebagai cara untuk melakukan pencocokan pola (pattern matching). Dalam struktur ini, setiap case memeriksa apakah nilai yang cocok dengan pola yang diberikan (misalnya, x, y, atau z), dan jika ada kecocokan, blok kode yang sesuai akan dieksekusi.

```python
day = 4
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
```

### Analisis
Kode tersebut menggunakan match-case untuk memeriksa nilai variabel day dan mencetak nama hari yang sesuai. Karena day bernilai 4, kondisi case 4 terpenuhi, dan program akan mencetak "Thursday".

### Default Value

```python
day = 4
match day:
  case 6:
    print("Today is Saturday")
  case 7:
    print("Today is Sunday")
  case _:
    print("Looking forward to the Weekend")
```

### Analisis
Kode tersebut menggunakan match-case untuk memeriksa nilai variabel day. Karena day bernilai 4, yang tidak cocok dengan case 6 atau case 7, program akan mencetak "Looking forward to the Weekend", karena kondisi case _ menangani semua kasus lainnya.

### Combine Values

```python
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5:
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
```

### Analisis
Kode tersebut menggunakan match-case dengan pola gabungan untuk memeriksa nilai variabel day. Karena day bernilai 4, yang termasuk dalam kelompok 1 | 2 | 3 | 4 | 5, program akan mencetak "Today is a weekday". Jika day bernilai 6 atau 7, maka program akan mencetak "I love weekends!".

### If Statements as Guards

```python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
```

### Analisis
Kode tersebut menggunakan match-case dengan kondisi tambahan yang memeriksa bulan (month). Karena day bernilai 4 dan month bernilai 5, kondisi kedua case 1 | 2 | 3 | 4 | 5 if month == 5 terpenuhi. Program akan mencetak "A weekday in May".

Jika month adalah 4 dan day adalah 4, maka "A weekday in April" akan dicetak, tetapi karena month adalah 5, program akan memilih kondisi kedua.
