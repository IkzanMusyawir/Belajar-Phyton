```python
# Impor pustaka Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data untuk visualisasi
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Buat plot
plt.plot(x, y, label='sin(x)')
plt.title('Contoh Plot Sederhana')
plt.xlabel('Nilai x')
plt.ylabel('Nilai y')
plt.legend()
plt.show()
```

```python
# Impor pustaka Pandas untuk bekerja dengan data tabular (DataFrame)
import pandas as pd

# Buat DataFrame dari dictionary
data = {'Nama': ['Budi', 'Siti', 'Rudi', 'Ani'],
        'Usia': [22, 25, 23, 26],
        'Kota': ['Jakarta', 'Bandung', 'Surabaya', 'Jakarta']}
df = pd.DataFrame(data)

# Tampilkan DataFrame
print("DataFrame Awal:")
print(df)

# Lakukan operasi sederhana, misalnya filter data
df_jakarta = df[df['Kota'] == 'Jakarta']

print("\nData penduduk dari Jakarta:")
print(df_jakarta)
```
